```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#importar archivo excel csv
surveys = pd.read_csv("surveys.csv")
```

```python
#generar grafica basico
my_plot = surveys.plot("hindfoot_length", 'weight',kind ="scatter") #ejex,ejey,tipo de grafica
```

```python
#Grafica del tipo histograma:
sample_data = np.random.normal(0,0.1,1000) #valor minimo, valor maximo, cantidad de datos a generar
sample_data
```

```python
#obteniendo histograma con los datos obtenidos:
plt.hist(sample_data)
plt.show() #se usa para obtener una grafica mas limpia
```

```python
#generar un archivo png de la grafica anterior
plt.savefig("histograma.png")
```

```python
#
x_values = surveys['year'].unique() #solo toma los valores una ves sin que se repitan
y_values = surveys['year'].value_counts().tolist() #conteo los valores de la columan year
```

```python
#obteniendo un grafico de barras.
plt.bar(x_values, y_values)

#para generar un titulo a la imagen
plt.title("Sobrevivientes por aÃ±o")

#edicion de las etiquetas del ejex y ejey
ax = plt.subplot()
ax.set_xlabel("AÃ±o de sobrevivientes")
ax.set_ylabel("Canttidad de sobrevivientes por aÃ±o")

#Configuracion de colores y tipo de fuente para la grafica:
plt.rcParams['font.family'] = "sans-serif" #cambiar fuente
plt.rcParams["xtick.color"] = "#EF1313"
plt.rcParams["ytick.color"] = "#EF1313"

plt.show() #grafico mas limpio
```
